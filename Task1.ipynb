```python
import xgboost as xgb
import numpy as np
```

```python
with open('airfoil_self_noise.dat') as f:
	x = [[float(xi) for xi in line.split()] for line in f]

y = np.array([[xi.pop(-1)] for xi in x])
x = np.array(x)

t = 1000
x1 = x[:t,:]
x2 = x[t:,:]
y1 = y[:t,:]
y2 = y[t:,:]
```

```python
train = xgb.DMatrix(x, label = y)
test = xgb.DMatrix(x2, label = y2)
```

```python
param = {'silent' : 1, 'objective' : 'reg:linear', 'booster' : 'gblinear', 'alpha' : 0, 'lambda' : 1e-6}
watchlist = [(test, 'eval'), (train, 'train')]
num_round = 300
```

```python
bst = xgb.train(param, train, num_round, watchlist)
```

```python
pred = bst.predict(test)
labels = test.get_label()
```

```python
#root mean squared error
np.sqrt(sum(map(lambda a,b: (a - b)**2 / x2.shape[0], pred, labels)))
```
